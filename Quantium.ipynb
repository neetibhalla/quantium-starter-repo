```python
import pandas as pd
df=pd.read_csv('data\\daily_sales_data_0.csv')
df.head()
df.info()
```

```python
df1=pd.read_csv('data\\daily_sales_data_1.csv')
df1.head()
```

```python
df2=pd.read_csv('data\\daily_sales_data_2.csv')
df2.head()
```

```python
df=df[df["product"]=="pink morsel"]
df.info()
```

```python
df1=df1[df1["product"]=="pink morsel"]
df2=df2[df2["product"]=="pink morsel"]
df1.info()
df2.info()
```

```python
df["price"] = df["price"].str.replace("$", "", regex=False).astype(float)
df["sales"] = df["price"] * df["quantity"]
```

```python
df1["price"] = df1["price"].str.replace("$", "", regex=False).astype(float)
df1["sales"] = df1["price"] * df["quantity"]
df2["price"] = df2["price"].str.replace("$", "", regex=False).astype(float)
df2["sales"] = df2["price"] * df["quantity"]
```

```python
columns_to_keep = ["product","sales", "region", "date"]

df = df[columns_to_keep]
df1 = df1[columns_to_keep]
df2 = df2[columns_to_keep]
```

```python
combined_df = pd.concat([df, df1, df2], ignore_index=True)
```

```python
combined_df.info()
```

```python
combined_df = combined_df.sort_values(by=["date", "region"])
```

```python
combined_df
```

```python
combined_df["date"] = pd.to_datetime(combined_df["date"])
combined_df.to_csv("output.csv", index=False)
```
